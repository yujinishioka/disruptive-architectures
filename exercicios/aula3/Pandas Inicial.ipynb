# **Introdução ao Pandas**

Pandas é uma ferramenta de manipulação de dados de alto nível, construída com base no pacote Numpy. 

Series: arrays unidimensionais rotulados capazes de armazenar qualquer tipo de dado.

Data Frames: estruturas de dados tabulares bidimensionais com rótulos nas linhas e colunas. Assim como as series, os dataframes são capazes de armazenar qualquer tipo de dado.
